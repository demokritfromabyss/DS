### Предобработка данных

#### Изменение регистра заголовков

```python
# форматируем названия столбцов, приводим к нижнему регистру
data.columns = data.columns.str.lower()
```

#### Изменение типа данных

```python
data['year_of_release'] = data['year_of_release'].astype('Int64')
```

Для столбца user_score не удалось скорректировать формат на float так как присутствуют значения tbd(означает to be determined), для корректного изменения формата предлагается заменить значения tbd на пропуски (NaN), так как подобная замена не повлияет на общую картину оценок, по сравнению с тем, если бы мы произвели замену на 0. 

```python
data.loc[data['user_score'] == 'tbd', 'user_score'] = np.NaN # заменяем tbd на NaN
print(data['user_score'].isna().sum()) # выводим число пропусков после замены
data['user_score'] = pd.to_numeric(data['user_score']) # изменяем тип данных
print(data.info()) # выводим общую информацию о датасете
```

#### Обработка пропусков

Пропущенные значения имеются в столбцах name, year_of_release, genre, critic_score, user_score, rating. 
* В столбцах name, year_of_release пне будем выполнять замену, т.к. отсутствуют подходящие для заполнения значения. Строки с пропущенными данными можем удалить, так как таких пропусков порядка 1,6% и на качестве данных это не отразится.
* Для столбцов critic_score, user_score оставляем пропуски, так как замены приведут к снижению качества данных.
* Для столбца rating выполним замену пропусков на значение NR (not rated), тем самым сделаем возможным объединение с группой, которая ожидает оценки. Также среди значений этого столбца есть К-А (от 6 лет и старше), в 1998 г. эта категория была изменена на Е.
* Пропуски в столбце genre будут исключены при удалении строк с пропусками в столбце name - так как эти строки не информативны.

Причины пропусков могут быть:
* полностью случайные (забыли указать жанр, например), 
* случайные (возможно оценки связаны связаны с платформой размещения игры, но есть более существенные факторы влияния на оценки), 
* неслучайные (рейтинг не был указан намеренно, чтобы расширить аудиторию). 
* технического характера: сбой в системе, некорректная выгрузка данных.

```python
print(data['rating'].value_counts())
```

```python
display(data.loc[data['genre'].isna()])
```

```python
# удалим строки с пропусками в названии
data = data.dropna(subset=['name']).reset_index(drop=True) 

# удалим строки с пропусками в столбце year_of_release
data = data.dropna(subset=['year_of_release']).reset_index(drop=True)

# заменим пропуски в рейтинге на NR
data['rating'] = data['rating'].fillna('NR')
data.loc[data['rating'] == 'RP', 'rating'] = 'NR'
print(data['rating'].isna().sum())

print(data.info())
```

```python
# создадим новый столбец с общим объемом продаж
data['total_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales']
```

```python
#выполняем проверку на наличие дубликатов или аномалий
print(data['name'].sort_values().unique())
print(data['genre'].sort_values().unique())
print(data['platform'].sort_values().unique())
print(data['year_of_release'].value_counts().sort_values())
print()

#выполняем проверку на неявные дубликаты
columns_to_check = ['name', 'platform', 'year_of_release']
duplicates = data.duplicated(subset=columns_to_check, keep=False)
duplicate_rows = data[duplicates]
print(duplicate_rows)
```

```python
#удалим неявные дубликаты и проверим что получилось
data = data.drop_duplicates(subset=columns_to_check)
data.info()
print(data.iloc[600])
```

#### Вывод

Выполнили предварительную обработку данных: изменен регистр, добавлены пропуски в тех местах, где это было возможно, и удалены строки с пропусками в тех случаях, когда их замена на иное значение привела бы к снижению качества данных. В случае с пользовательской оценкой, выполнили замену на пропуски, для корректности дальнейшего анализа. Был добавлен столбец с общим объемом продаж по регионам, также выполнили проверку на наличие дубликатов для столбцов с платформами и жанрами.

### Исследовательский анализ данных

#### Динамика объемов выпускаемых игр

```python
data.groupby('year_of_release')['name'].count().plot(kind='bar', figsize=(10,5), color='darkblue', grid=True)
plt.title('Количество игр, выпущенных в разные годы')
plt.xlabel('Год выпуска')
plt.ylabel('Количество игр')
plt.show()

data.groupby('year_of_release')['total_sales'].sum().plot(kind='bar', figsize=(10,5), color='red', grid=True)
plt.title('Годовые объемы продаж')
plt.xlabel('Год выпуска')
plt.ylabel('Объем продаж')
plt.show()
```

**Вывод:** по графику видно, что до 1994 г. количество выпускаемых игр было крайне мало, и ими можно пренебречь. С 1994 по 2009 годы, прослеживается положительная динамика количества выпускаемых игр, однако носит неравномерный характер: с 2002 г. происходит 2-3 кратный рост, начиная с 2010 г. тенденция снижается. Так как восстановления количественного показателя выпускаемых игр не произошло, выберем актуальным период с 2014 по 2016 г.

#### Динамика продаж в разрезе платформ

```python
platform_sales = data \
                .pivot_table(index='platform', values='total_sales', aggfunc='sum') \
                .sort_values(by='total_sales', ascending=False)
platform_sales.plot(kind='bar', figsize=(10,5), color='darkgreen', grid=True)
plt.title('Объем продаж в разрезе платформ')
plt.xlabel('Платформа')
plt.ylabel('Общий объем продаж')
plt.show()
```

**Вывод:** На платформы PS2, X360, PS3, Wii, DS, PS приходится Топ-6 продаж, их объемы в 3 и более раза превышают значения по другим платформам. Проверим, сохраняется ли данная тенденция в выбранный период 2014-2016гг.

```python
data.plot(kind='scatter', x='platform', y='year_of_release', figsize=(15,5), alpha=0.04, color='green')
plt.title('Популярность платформы в зависимости от года')
plt.xlabel('Платформа')
plt.ylabel('Год выпуска')
plt.show()
data \
    .query('2012 <= year_of_release <= 2016') \
    .pivot_table(index='year_of_release', columns='platform', values='total_sales', aggfunc='sum') \
    .plot(figsize=(10,5), grid=True)
plt.title('Продажи игр за 2012-2016гг.')
plt.xlabel('Год выпуска')
plt.ylabel('Объем продаж')
plt.show()

sales_top = data.query('year_of_release >= 2014').groupby('platform')['total_sales'].sum()
display(sales_top.sort_values(ascending=False))
```

**Вывод:** 
из ранее выявленных платформ лидеров по продажам, при анализе по годам, в выборку 2012-2016 гг. не попали PS, PS2, DS (c 2013г.), что говорит о том, что они были популярны в предыдущие годы, а с 2013 года на рынок вышли новые платформы - PS4 и XOne. В среднем, по данным графиков, жизненный цикл платформы составляет порядка 10 лет, при этом есть исключения, например платформа РС, жизненный цикл превышает 20 лет, а также нежизнеспособные платформы (N64, SAT, DC). В среднем в течение жизненного цикла платформы около 5-7 лет приходится на активные продажи, и далее рост объема продаж на них замедляется или объемы продаж на них снижаются. Соответственно выделим следующие категории платформ:
* лидеры индустрии - PS4, XOne, 3DS, которые в 2016г. входят в топ-5 платформ по продажам;
* бывшие лидеры индустрии - X360, PS3, вышедшие в 2005г. и сохраняющие лидирующие позиции по объемам продаж среди платформ;
* платформа PSV может стать потенциально прибыльной - жизненный цикл платформы еще не завершен;
* платформы, ушедшие с рынка к 2016г. или демонстрирующие существенную тенденцию к снижению - DS, PSP, Wii.

Динамика объема продаж по всем платформам демонстрирует существенное снижение за рассматриваемый период с 2014 по 2016 гг.

#### Диаграмма размаха продаж игр на платформах

```python
data_for_boxplot = data.query('year_of_release >= 2014')

plt.figure(figsize=(15,10))
sns.boxplot(data=data_for_boxplot, y='total_sales', x='platform', color='cornflowerblue')
plt.ylim(0, 2)
plt.title('Сравнение платформ', fontsize = 18)
plt.xlabel('Платформа')
plt.ylabel('Объем продаж')
plt.show()

plt.figure(figsize=(15,10))
sns.boxplot(data=data_for_boxplot, y='total_sales', x='platform', color='cornflowerblue')
plt.ylim(0, 16)
plt.title('Сравнение платформ', fontsize = 18)
plt.xlabel('Платформа')
plt.ylabel('Объем продаж')
plt.show()

display(data_for_boxplot.pivot_table(index='platform', values='total_sales', aggfunc='describe'))
```

**Вывод:** построив диаграммы размаха по платформам, актуальным в 2014-2016 гг., видим, что наибольшее число выбросов характерно для PS4, 3DS, XOne, PS3, что, скорее всего, связано с выходом эксклюзивных игр на этих платформах. Максимальные значения у платформ PS4 и 3DS - 14.6 и 11.7 соответственно при среднем 0.8 и 0.4 соответственно, также у этих платформ наибольшее стандартное отклонение. Так как медиана не смещена в сторону выбросов, можем сделать вывод что число выбросов невелико относительно нормальных значений. 

#### Влияние на продажи игр отзывов критиков и пользователей

```python
example = data_for_boxplot.query('platform == "PS4"') # создали датафрейм для анализа PS4 - по этой платформе наибольшие продажи
display(example)
```

```python
# определяем сетку и оси с помощью matplotlib.pyplot.subplots()
fig, ax = plt.subplots(1, 2, figsize=[15, 5])

fig.suptitle('Влияние на продажи игр оценок', fontsize=16)

# строим графики
ax[0].scatter(example['critic_score'], example['total_sales'], color='coral')
ax[0].set_title('критиков', fontsize=15, fontstyle='italic')
ax[0].set_xlabel('Оценка', fontsize=11, fontstyle='italic')
ax[0].set_ylabel('Продажи, млн.', fontsize=11, fontstyle='italic')

ax[1].scatter(example['user_score'], example['total_sales'], color='cadetblue')
ax[1].set_title('пользователей', fontsize=15, fontstyle='italic')
ax[1].set_xlabel('Оценка', fontsize=11, fontstyle='italic')
ax[1].set_ylabel('Продажи, млн.', fontsize=11, fontstyle='italic')

plt.show()
```

```python
# посчитаем корреляцию между объемами продаж и оценкой критиков
print(example['critic_score'].corr(example['total_sales']))

# посчитаем корреляцию между объемами продаж и оценкой пользователей
print(example['user_score'].corr(example['total_sales']))
```

**Вывод:**

### Проверка гипотез

#### Сравнение пользовательских рейтингов платформ Xbox One и PC

**Н0:** средняя оценка пользователей для платформы Xbox One равна средней оценке пользователей для платформы PC.

**Н1:** средние оценки пользователей для платформ Xbox One и PC не равны.


Это двусторонняя гипотеза, генеральные совокупности не зависят друг от друга (оценки пользователей одной платформы не зависят от оценок пользователей другой платформы). Уровень статистической значимости примем 0.05, мы не знаем равны ли дисперсии генеральных выборок, поэтому примем equal_var = False (но это можно преодолеть благодаря ЦПТ). Добавим параметр nan_policy='omit', т.к. без него p-значение не рассчитывалось - выводится nan. 

```python
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(
    data_for_boxplot[data_for_boxplot['platform'] == 'XOne']['user_score'], 
    data_for_boxplot[data_for_boxplot['platform'] == 'PC']['user_score'],
    equal_var = False, nan_policy='omit')

# alpha = задайте значение уровня значимости
alpha = 0.05

# вывод значения p-value на экран
print('p-значение:', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
    
print()

# посчитаем средние оценки пользователей
xone_users = data_for_boxplot[data_for_boxplot['platform'] == 'XOne']['user_score']
print('Средняя пользовательская оценка для платформы XOne:', xone_users.mean())

pc_users = data_for_boxplot[data_for_boxplot['platform'] == 'PC']['user_score']
print('Средняя пользовательская оценка для платформы PC:', pc_users.mean())   
```

**Вывод:** по результатам проверки, делаем вывод, что нулевая гипотеза о равенстве оценок пользователей платформ Xbox One и PC не может быть отвергнута.

#### Сравнение пользовательских рейтингов жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования»)

**Н0:** пользовательские рейтинги игр в жанре "экшен" и "спорт" равны.

**Н1:** пользовательские рейтинги игр в жанре "экшен" и "спорт" не равны.


Это двусторонняя гипотеза, генеральные совокупности не зависят друг от друга (оценки пользователей одного жанра игр не зависят от оценок пользователей другого жанра игр). Уровень статистической значимости примем 0.05, однако мы не знаем равны ли дисперсии генеральных выборок, поэтому примем equal_var = False (но это можно преодолеть благодаря ЦПТ). Добавим параметр nan_policy='omit', т.к. без него p-значение не рассчитывалось - выводится nan. 

```python
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(
    data_for_boxplot[data_for_boxplot['genre'] == 'Action']['user_score'], 
    data_for_boxplot[data_for_boxplot['genre'] == 'Sports']['user_score'],
    equal_var = False, nan_policy='omit')

# alpha = задайте значение уровня значимости
alpha = 0.05

# вывод значения p-value на экран
print('p-значение:', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
    
print()

# посчитаем средние оценки пользователей
action_users = data_for_boxplot[data_for_boxplot['genre'] == 'Action']['user_score']
print('Средняя пользовательская оценка для игр жанра Action:', action_users.mean())

sports_users = data_for_boxplot[data_for_boxplot['genre'] == 'Sports']['user_score']
print('Средняя пользовательская оценка для игр жанра Sports:', sports_users.mean())        
```

**Вывод:** по результатам проверки гипотезы делаем вывод, что p-значение близко к нулю, отвергаем нулевую гипотезу о равенстве оценок пользователей игр жанра Action и Sports.
